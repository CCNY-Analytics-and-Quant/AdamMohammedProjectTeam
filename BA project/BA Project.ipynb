## 3. Targeted Promotions and discounts & sales for Top Products and Customers In Top Selling Seasons :

#### Customer Purchase Patterns:
we have Analyzed purchase history to identify trends and preferences of top customers and products, and we have identified three seasons in a year when most sales happen, we can offer discounts during this time

```python
import pandas as pd
import matplotlib.pyplot as plt

# list to store individual customer purchase patterns
purchase_patterns = []

# Loop through the top customers
for customer_id in top_customers.index:
    # Filter order data for the selected customer
    customer_orders = order_data[order_data['Order_Customer_Id'] == customer_id]

    # Convert 'Order_Date' to datetime format using .loc
    customer_orders.loc[:, 'Order_Date'] = pd.to_datetime(customer_orders['Order_Date'])

    # Calculate purchase frequency by month
    purchase_frequency = customer_orders['Order_Date'].value_counts().resample('M').count()

    # Append the purchase frequency data to the list
    purchase_patterns.append({'Customer_ID': customer_id, 'Purchase_Frequency': purchase_frequency})

# line charts for purchase frequency over time for each top customer
plt.figure(figsize=(12, 6))
for pattern in purchase_patterns:
    plt.plot(pattern['Purchase_Frequency'].index, pattern['Purchase_Frequency'].values, marker='o', label=f"Customer {pattern['Customer_ID']}")

plt.xlabel('Month')
plt.ylabel('Purchase Frequency')
plt.title('Purchase Frequency Over Time for Top Customers')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
